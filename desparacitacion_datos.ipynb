```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Cargar datos para revisar la correlación de variables, cuentas y demas

```python
df = pd.read_csv('data/Datos_proyecto_II_BI_2021.csv', sep=',', encoding='latin-1')
df.shape
```

```python
df.head()
```

```python
desparacitacion = df["NPCFP7A1"]
```

```python
# Count null in desparacitacion
desparacitacion.isnull().sum()
```

```python
# Fill null values with 0 in column NPCFP7A1
df["NPCFP7A1"] = df["NPCFP7A1"].fillna(0)

# Replace 1.0 with yes and 2.0 with no
df["NPCFP7A1"] = df["NPCFP7A1"].replace(1.0, "Yes")
df["NPCFP7A1"] = df["NPCFP7A1"].replace(2.0, "No")
```

```python
# Graph percentage of desparacitacion in pie graph
desparacitacion = df["NPCFP7A1"]
desparacitacion = desparacitacion.value_counts(normalize=True) * 100
desparacitacion.plot.pie(autopct="%.1f%%")
plt.title("Desparacitacion - 2021")
plt.show()
```

```python
# Load csv 2017
df_2017 = pd.read_csv('data/Datos_proyecto_II_BI_2017.csv', sep=',', encoding='latin-1')
```

```python
# Fill null values with 0 in column NPCFP7A1
df_2017["NPCFP33E"] = df_2017["NPCFP33E"].fillna(0)

# Replace 1.0 with yes and 2.0 with no
df_2017["NPCFP33E"] = df_2017["NPCFP33E"].replace(1.0, "Yes")
df_2017["NPCFP33E"] = df_2017["NPCFP33E"].replace(2.0, "No")
```

```python
# Graph percentage of desparacitacion in pie graph
desparacitacion_2017 = df_2017["NPCFP33E"]
desparacitacion_2017 = desparacitacion_2017.value_counts(normalize=True) * 100
desparacitacion_2017.plot.pie(autopct="%.1f%%")
plt.title("Desparacitacion - 2017")
plt.show()
```
